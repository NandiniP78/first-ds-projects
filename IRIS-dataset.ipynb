```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = sns.load_dataset('iris')
df
```

```python
#pairplot
#change style and context
sns.set_style('darkgrid')
sns.set_context('talk')
sns.pairplot(df,hue="species")
```

```python
#scatterplot
sns.scatterplot(data=df,x="sepal_length",y="petal_length",hue='species')
```

```python
sns.displot(data=df,x='sepal_length',hue='species',bins=20)
```

```python
sns.violinplot(data=df,x='species',y='sepal_width')
```

```python
sns.kdeplot(data=df,x="petal_length",y="petal_width")
```

```python
sns.boxplot(data=df,x='species',y='petal_length')
plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length(cm)')
plt.figure(figsize=(8,6))
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
```

```python
sns.set_palette('Set2')
sns.barplot(data=df,x='species',y='petal_length')
```

```python
sns.countplot(data=df,x='species')
```

```python
sns.regplot(data=df,x='sepal_width',y='petal_width')
```

```python
g = sns.FacetGrid(data=df,col='species')
g.map(sns.scatterplot,'sepal_length','petal_length')
```

```python
g = sns.FacetGrid(data=df,col='species')
g.map(sns.histplot,'sepal_length')
```

```python
#Use a pairplot to display the relationships between all the numerical features, but only for the species setosa (filter the dataset)
setosa = df[df['species'] == 'setosa']

sns.pairplot(setosa,hue='species')
```

```python
#regression plot
sns.regplot(data=df,x='sepal_length',y='petal_length')
```

```python
#Use Seabornâ€™s lmplot to perform a linear regression on sepal_length and sepal_width, and color the plot by species.
#linear regression plot 
sns.lmplot(data=df,x='sepal_length',y='sepal_width',hue='species')
```

```python
# Introduce missing data by randomly removing some values
iris_missing = df.copy()
iris_missing.loc[iris_missing.sample(frac=0.1).index, 'sepal_length'] = None

# Visualizing missing data
sns.heatmap(iris_missing.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
# Scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', data=df)
plt.title('Sepal Length vs Petal Length')

# Add annotation to a point
plt.annotate('Important Point', xy=(5, 1.5), xytext=(6, 3), 
             arrowprops=dict(arrowstyle='->'))
plt.show()
```

```python
# Jointplot with scatter kind
sns.jointplot(x='sepal_length', y='petal_length', data=df, kind='scatter')
plt.show()

# Jointplot with hexbin kind
sns.jointplot(x='sepal_length', y='petal_length', data=df, kind='hex')
plt.show()
```
